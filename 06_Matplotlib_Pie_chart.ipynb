## Matplotlib Pie Chart
    A pie chart in Matplotlib is used to show how a whole is divided into parts.Each slice represents a categoryâ€™s proportion relative to the total. The matplotkib plt.pie() function help to plot pie chart of given numeric data with labels. 
    Purpose of Pie Chart ::
    - Display percentage or proportion
    - Compare parts of a whole

```python
import pandas as pd
import numpy as np
import random 
import matplotlib.pyplot as plt 
import matplotlib.style as style
style.use("ggplot")
```

### Example - Basic Pie chart 

```python
# Plotting Matplotlib Pie Chart
plt.pie([1]) # Plot pie chart of value [1]
plt.show() 
```

### Example - Percentage of students in each class

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence','Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students, labels = classes,colors = colors)
plt.show()
```

```python
explode = [0,0,0.1,0,0]         # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":15}     # Font size of text in pie chart
plt.pie(class1_students,labels = classes,explode=explode,colors = colors,textprops = textprops, autopct="%0.1f%%",shadow=True,startangle=180)
# explode = To slice the perticuler section
# autopct = Show data in persentage for with 2 decimal point
# Radius = to increase or decrease the size of pie chart
plt.show()
```

### Example - Stock Portfolio Allocation

```python
# Stock names
stocks = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tesla']
investment = [30, 25, 20, 15, 10]  # Investment amount (or percentage)
explode = (0.1, 0, 0, 0, 0)  # Optional: explode the largest stock

# Create pie chart
plt.pie(investment,labels=stocks,autopct='%1.1f%%',startangle=90,colors = colors, explode=explode, shadow=True)

# Title
plt.title("Stock Portfolio Allocation")
plt.show()
```
