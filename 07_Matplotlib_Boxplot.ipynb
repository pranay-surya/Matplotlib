## Boxplot 
    A Matplotlib boxplot [ a.k.a. box-and-whisker plot ] is a compact way to summarize the distribution of numerical data—showing spread, center, and outliers at a glance.
    
    What a boxplot shows ::
    - Median (Q2): The line inside the box → the middle value.
    - Quartiles (Q1 & Q3): The bottom and top of the box → middle 50% of data.
    - Interquartile Range (IQR): Q3 − Q1 → spread of the central data.
    - Whiskers: Extend to the smallest/largest values within 1.5 × IQR.
    - Outliers: Points beyond the whiskers → unusually high/low values.

    Why use a boxplot?
    - Quickly compare distributions across groups.
    - Detect outliers.
    - Understand skewness and variability without assuming normality.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.style as style 
style.use("ggplot")
```

### Example - Basic Boxplot 

```python
# random data samples
np.random.seed(10)
data = np.random.normal(50,10,100)  # mean = 50, std = 10, n = 100 
plt.boxplot(data)                   # create box plot
plt.title('Boxplot Example')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.show()
# Box Middle line --> median (nearest 50)
# Box Botton side --> Q1 (25th percentile)
# Box Top side --> Q3 (75th percentile)
# Outer points --> Outliers
```

### Example - Differance of Histogram vs Boxplot in Code


```python
plt.figure(figsize=(15,5))  # Create a figure with a wide layout 

# -------------------- Histogram --------------------
plt.subplot(1,2,1)  # Create the first subplot 
np.random.seed(10)

# Generate 100 random values from a normal distribution
# mean = 50, standard deviation = 10
data = np.random.normal(50, 10, 100)

# Plot histogram with 20 bins and black edges
plt.hist(data, bins=20, edgecolor='black')

# Add title and axis labels
plt.title('Histogram Example')
plt.xlabel('Dataset')
plt.ylabel('Values')


# -------------------- Boxplot --------------------
# Create the second subplot (1 row, 2 columns, position 2)
plt.subplot(1,2,2)

# Plot vertical boxplot for the same data
plt.boxplot(data, vert=True)

# Add title and axis labels
plt.title('Boxplot Example')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.show()
```

### Example - Spread and Outlier Detection Across Datasets

```python
# Fix seed for reproducibility
np.random.seed(42)
data1 = np.random.normal(50, 10, 100)   # Dataset A
data2 = np.random.normal(60, 15, 100)   # Dataset B
data3 = np.random.normal(55, 8, 100)    # Dataset C
plt.figure(figsize=(15,5))  # Create a figure with 1 row and 3 columns

# -------- Boxplot 1 --------
plt.subplot(1, 3, 1)
plt.boxplot(data1)
plt.title("Boxplot - Dataset A")
plt.ylabel("Values")

# -------- Boxplot 2 --------
plt.subplot(1, 3, 2)
plt.boxplot(data2)
plt.title("Boxplot - Dataset B")

# -------- Boxplot 3 --------
plt.subplot(1, 3, 3)
plt.boxplot(data3)
plt.title("Boxplot - Dataset C")
plt.tight_layout()
plt.show()
```
