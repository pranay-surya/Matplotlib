## Matplotlib Bar Chart
    A bar chart in Matplotlib is used to visualize and compare categorical data by representing values as rectangular bars, where the height (or length) of each bar corresponds to the data value.
    Purpose of a Bar Chart ::
     - Compare values across different categories
     - Identify highest and lowest values easily
     - Present data in a clear and readable format

```python
import pandas as pd
import numpy as np
import random 
import matplotlib.pyplot as plt
import matplotlib.style as style 
style.use('ggplot')
```

### Generating Sample Data for Plotting 

```python
# Importing Data set to plot Bar Chart
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

### Example - Number of students in programming language  

```python
plt.bar(classes, class1_students,width=0.5,align = 'center',edgecolor = "black", linewidth = 1,alpha = 0.7,label = 'Class_1_students')  
# Plot vertical Bar Chart
# for x-axis = Classes
# for y-axis = classes1_students
# align : {‘center’, ‘edge’}, optional, ……….default: ‘center’
# alpha :: Transferncy of bars 
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Programming Lang. students",fontdict = font1)
plt.xlabel("Languages",fontdict= font2)
plt.ylabel("No. of Students",fontdict  = font2)
plt.legend()
plt.show()
```

### Plotting for multiple classes 

```python
# Create an array of indices based on the number of classes (categories)
classes_index = np.arange(len(classes))  
index = np.arange(len(classes))
width = 0.2 # Define the width of each bar

# Plot bar chart for Class 1 students
# Bars are placed at the original index positions
plt.bar(classes_index, class1_students, width,color="b", label="Class 1 Students")  

# Plot bar chart for Class 2 students
# Bars are shifted to the right by 'width' to avoid overlap
plt.bar(classes_index + width, class2_students, width, color="g", label="Class 2 Students")

# Plot bar chart for Class 3 students
# Bars are shifted further right by '2 * width'
plt.bar(classes_index + width + width, class3_students, width,color="y", label="Class 3 Students")

# Set the title and label of the bar chart
plt.title("Programming Lang. students", fontdict=font1)
plt.xlabel("Languages", fontdict=font2)
plt.ylabel("No. of Students", fontdict=font2)
plt.xticks(index + width,classes)
plt.legend()  # Display legend to identify each class
plt.show()
```

### Example - Daily Trading Volume 

```python
stocks = ['TCS', 'Infosys', 'Reliance', 'HDFC', 'Wipro']  # Stock names
volume = [120, 95, 150, 110, 80] # Stock Volume ( in thousands )

plt.bar(stocks, volume, color='steelblue')
plt.title('Daily Trading Volume of Stocks')
plt.xlabel('Stock Name')
plt.ylabel('Volume ( in thousands )')
plt.show()
```

### Example - Campnies Stock Camparision 

```python
stocks = ['TCS', 'Infosys', 'Reliance','Tech Mahindra']  # Stock names
 
# Stock prices
open_price = [3400, 1500, 2450,2800]
close_price = [3450, 1520, 2500,3000]
high_price = [3480, 1540, 2520,3100]

# Create index for stocks
group_gap = 0.5
index = np.arange(len(stocks)) * (3 * 0.25 + group_gap)

# Width of each bar
width = 0.2

# Plot multiple bars side by side
plt.bar(index, open_price, width, label='Open Price')
plt.bar(index + width, close_price, width, label='Close Price')
plt.bar(index + 2*width, high_price, width, label='High Price')

# Labels and title
plt.xlabel('Stocks')
plt.ylabel('Price')
plt.title('Stock Price Comparison')
plt.xticks(index + width, stocks)
plt.legend()
plt.show()
```
