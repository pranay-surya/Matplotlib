## Matplotlib Scatter Plot
    A scatter plot in Matplotlib is used to visualize the relationship between two numerical variables by displaying data points on a 2D plane. Each point represents one observation with an (x, y) coordinate
    Basic Purpose ::
    - Show how one variable changes with another
    - Identify patterns, trends, clusters, or outliers

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.style as style 
style.use("ggplot")
```

```python
data = pd.read_csv("googleplaystore.csv")
data.head(2)
```

```python
data = pd.read_csv("googleplaystore.csv",nrows=1000)
data.shape
```

```python
font1 = {
    'family': 'serif',
    'color': 'darkred',
    'size': 18
}
font2 = {
    'family': 'serif',
    'color': 'darkred',
    'size': 15
}
```

### Example - Correlation Between Ratings and Reviews

```python
x = data['Rating']   # Select the 'Rating' column for X-axis values
y = data['Reviews']  # Select the 'Reviews' column for Y-axis values

# Create a scatter plot
plt.scatter(x, y, marker='^', edgecolor='black', alpha=0.6, color='darkred')

# Set the title and label of the scatter plot
plt.title("Google Play Store Apps Scatter plot", fontdict=font1)
plt.xlabel("Rating", fontdict=font2)
plt.ylabel("Reviews", fontdict=font2)
plt.show()
```

### Example - Google Play Store Apps: Rating vs Reviews & Installs

```python
# For plotting two scatter plots on a single graph
x = data['Rating']    # X-axis values: App ratings
y = data['Reviews']   # Y-axis values for first scatter plot: Number of reviews

# Scatter plot for Reviews
plt.scatter(x, y, marker='^', edgecolor='darkred',alpha=0.5,color='darkred', label='Reviews')

# Scatter plot for Installs (same X-axis, different Y-axis)
plt.scatter(x, data['Installs'],marker='o',edgecolor='black',alpha=0.5,color='black',label='Installs')

# Set the title and label of the plot
plt.title("Google Play Store Apps Scatter plot", fontdict=font1)
plt.xlabel("Rating", fontdict=font2)
plt.ylabel("Reviews & Installs", fontdict=font2)
plt.legend(loc=2)
plt.show()
```
