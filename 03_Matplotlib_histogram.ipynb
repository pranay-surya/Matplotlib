## Python Matplotlib Histogram
    A histogram is a graphical representation used to show the distribution of numerical (continuous) data.
    It groups data into ranges (bins) and displays how frequently values fall into each range.
    It form a rectangle bar,each bar shows some data, which belong to different categories. 
    A histogram helps you understand ::
    - Data distribution (normal, skewed, uniform)
    - Frequency of values
    - Spread and concentration of data
    - Presence of outliers

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import matplotlib.style as style
import random
style.use("ggplot")
```

### Sample Data for Histogram 

```python
ml_students_age = np.random.randint(18,50, (100))
py_students_age = np.random.randint(15,50, (100))
print(ml_students_age)
print(py_students_age)
```

### Font Style 

```python
font1 = {'family':'serif','color':'darkred','size':18} 
font2 = {'family':'serif','color':'darkred','size':15}
```

### Example - Plotting histogram of machine learning students age.

```python
# Plot histogram for ML students' age distribution
plt.hist(
    ml_students_age,                 # List/array of student ages
    label="ML_Student",               # Label for the legend
    color="darkred",                  # Color of histogram bars
    rwidth=0.8,                       # Relative width of bars (gap between bars)
    bins=[15,20,25,30,35,40,45,50]    # Custom age intervals (bins)
)

# Set title and label of the histogram 
plt.title("ML Students age histograms", fontdict=font1)
plt.xlabel("Students age category", fontdict=font2) 
plt.ylabel("Students age", fontdict=font2)
plt.legend(loc=2) # Display legend at upper-left position
plt.show()
```

### Example - Histograms for ML and Python students together

```python
# Define age intervals (bins) for the histogram
bins = [15, 20, 25, 30, 35, 40, 45, 50]

# Plot histograms for ML and Python students together
plt.hist(
    [ml_students_age, py_students_age],  # Two datasets for comparison
    bins,                                # age bins
    rwidth=0.8,                          # Width of bars relative to bin width
    histtype="bar",                      # Bar-style histogram
    orientation='vertical',              # Vertical bars
    color=["black", "darkred"],          # Colors for each group
    label=["ML Student", "Py Student"]   # Labels for legend
)

# Set title and label for histogram 
plt.title("ML & PY Students age histograms", fontdict=font1)
plt.xlabel("Students age category", fontdict=font2)
plt.ylabel("Students age", fontdict=font2)
plt.legend(loc=2) # Display legend at upper-left position
plt.show()
```

### Example - Histogram of Stock Daily Returns

```python
# Daily percentage returns of a stock
daily_returns = [-2.5, -1.8, -1.2, -0.5, 0.2, 0.8, 1.1, 1.5, 2.0, 3.0,4.5]

# Plot histogram of stock returns
plt.hist(
    daily_returns,          # Numerical data (daily returns)
    bins=6,                  # Number of bins (intervals)
    rwidth=0.6,             
    color='darkgreen',      # Color of histogram bars
    edgecolor='black'      # Border color of bars
)

plt.title("Stock Daily Return Distribution", fontdict=font1)
plt.xlabel("Daily Return (%)", fontdict=font2)
plt.ylabel("Frequency", fontdict=font2)
plt.show()
```
